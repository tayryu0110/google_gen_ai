# Data Tranformation using Gemini
* Using few-shot examples for data cleaning and transformation
* Run on a single row to verify the function works as expected

```python
# Using few-shot examples for data cleaning and transformation
# Run on a single row to verify the function works as expected

few_shot_examples = [
    {
        "input": {
            "Name": "Bobby JacksOn",
            "Age": 30,
            "Gender": "Male",
            "Medical Condition": "Cancer",
            "Doctor": "Matthew Smith",
            "Billing Amount": 18856.28
        },
        "output": {
            "name": "Bobby Jackson",
            "age": 30,
            "gender": "Male",
            "condition_category": "Oncology",
            "assigned_doctor": "Dr. Matthew Smith",
            "billing_usd": 18856.28
        }
    },
    {
        "input": {
            "Name": "LesLie TErRy",
            "Age": 62,
            "Gender": "Male",
            "Medical Condition": "Obesity",
            "Doctor": "Samantha Davies",
            "Billing Amount": 33643.32
        },
        "output": {
            "name": "Leslie Terry",
            "age": 62,
            "gender": "Male",
            "condition_category": "Chronic",
            "assigned_doctor": "Dr. Samantha Davies",
            "billing_usd": 33643.32
        }
    }
]

# Convert examples into prompt
example_prompt = "\n\n".join([
    f"Input: {ex['input']}\nOutput (JSON): {ex['output']}"
    for ex in few_shot_examples
])

patient_record = df.iloc[2][['Name', 'Age', 'Gender', 'Medical Condition', 'Doctor', 'Billing Amount']].to_dict()

# Create the prompt with new input
final_prompt = example_prompt + f"\n\nInput: {patient_record}\nOutput (JSON):"

response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=final_prompt)

#print(response.text)
```

```python
# Use only TOP 10 rows in order to avoid ClientError: 429 RESOURCE_EXHAUSTED, since this is running on the free tier of the Gemini API.
df_sample = df.head(10)

result_df=[]
for i, row in df_sample.iterrows():
    response = client.models.generate_content(
    model="gemini-2.0-flash",
    contents= example_prompt + f"\n\nInput: {row}\nOutput (JSON):")
    result_df.append(response.text)
```

```python
len(result_df)
```

```python
from chromadb import Documents, EmbeddingFunction, Embeddings
from google.api_core import retry

from google.genai import types


# Define a helper to retry when per-minute quota is reached.
is_retriable = lambda e: (isinstance(e, genai.errors.APIError) and e.code in {429, 503})


class GeminiEmbeddingFunction(EmbeddingFunction):
    # Specify whether to generate embeddings for documents, or queries
    document_mode = True

    @retry.Retry(predicate=is_retriable)
    def __call__(self, input: Documents) -> Embeddings:
        if self.document_mode:
            embedding_task = "retrieval_document"
        else:
            embedding_task = "retrieval_query"

        response = client.models.embed_content(
            model="models/text-embedding-004",
            contents=input,
            config=types.EmbedContentConfig(
                task_type=embedding_task,
            ),
        )
        return [e.values for e in response.embeddings]
```

```python
documents = result_df
```

```python
import chromadb

DB_NAME = "googlecardb"
embed_fn = GeminiEmbeddingFunction()
embed_fn.document_mode = True  # document embedding mode

# Setup ChromaDB
chroma_client = chromadb.Client()
db = chroma_client.get_or_create_collection(
    name=DB_NAME,
    embedding_function=embed_fn
)


db.add(
    documents=documents,
    ids=[str(i) for i in range(len(documents))]
)
```

```python
db.count()
```

# RAG Model
* Retrival(Vector Search) Part - Subsetting Top 3 Female Patient
* Prompting 
* Prompt Engineering

```python
# Retrieval(Vector Search) - Subsetting Top 3 Female Patient

embed_fn.document_mode = False
query =""" 'gender':'Female', """
result = db.query(query_texts=[query], n_results=3)
[all_passages] = result["documents"]
```

```python
# Prompting
prompt = f"""You are a helpful and informative assistant that answers questions using the reference passages below.

Can you tell me who the assigned doctors are for all the patients?
"""

# Add retrieved passages
for passage in all_passages:
    passage_oneline = passage.replace("\n", " ")
    prompt += f"PASSAGE: {passage_oneline}\n"
```

```python
answer = client.models.generate_content(
    model="gemini-2.0-flash",
    contents=prompt)

print(answer.text)
```
